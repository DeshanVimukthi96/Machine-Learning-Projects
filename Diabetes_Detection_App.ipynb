```python
x=data.iloc[:,:8].values
y=data.iloc[:,8].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=RandomForestClassifier(n_estimators=500)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
st.subheader("Accuracy Score")
```

```python
st.write(accuracy_score(y_test,y_pred))
```

```python
def User_input():
    preg=st.slider("Pregnancies",0,20,0)
    glu=st.slider("Glucose",0,200,0)
    bp=st.slider("BloodPressure",0,140,0)
    sthick=st.slider("SkinThickness",0,100,0)
    ins=st.slider("Insulin",0.0,1000.0,0.0)
    bmi=st.slider("BMI",0.0,70.0,0.0)
    dpf=st.slider("DiabetesPedigreeFunction",0.000,3.000,0.000)
    age=st.slider("Age",0,120,0)

    input_dict={"Pregnancies":preg,"Glucose":glu,"BloodPressure":bp,"SkinThickness":sthick,"Insulin":ins,"BMI":bmi,"DiabetesPedigreeFunction":dpf,"Age":age}
    return pd.DataFrame(input_dict,index=["User Input Values"])
```
